# Cargar el dataset
Dado por la calidad del aire en puebla en excel

```python
# Importart librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Cargar el dataset de excel
data = pd.read_excel('data/csv/Datos 2022 2024p.xlsx', sheet_name='Estaciones de Monitoreo')
data.head()
```

```python
# Acomodar el data set a un formato que pueda ser leido
# remove the first column
data_clean = data.drop(data.columns[0], axis=1)
data_clean.head()
```

```python
# Poner nombre a las columnas realeales de la row 0
data_clean.columns = data_clean.iloc[0]
data_clean = data_clean[1:]
data_clean.head()
```

```python
# Eliminar columnas que no se van a usar
data_clean = data_clean.drop(['Equipo Analizador', 'Tipo', 'N.'], axis=1)


# Verifica el resultado
data_clean.head()
```

```python
# Borrar valores nulos
data_clean = data_clean.dropna()
data_clean.head()
```

```python
# Guardar estaciones limpias
data_clean.to_csv('data/csv/estaciones.csv', index=False)
```

# Obtener datos por year

```python
df = pd.read_excel('data/csv/Datos 2022 2024p.xlsx', sheet_name='2024 p')
df.head()
```

```python
df.columns = df.iloc[1]
df = df.drop(1)

# Renombrar la primera columna
df.columns.values[0] = 'DateTime'
df.head()
```

```python
# Lista de estaciones que deseas procesar
estaciones = ['AGUA SANTA', 'BINE', 'NINFAS', 'UTP', 'VELODROMO']

# Lista de contaminantes en orden esperado
nombres_columnas = ['DateTime', 'O3', 'CO', 'NO2', 'SO2', 'PM10', 'PM2.5', 'Estacion']

# DataFrame final que contendrá todas las estaciones
df_total = pd.DataFrame()

# Recorrer cada estación
for estacion in estaciones:
    # Seleccionar columnas: DateTime y la estación específica
    df_estacion = df[['DateTime', estacion]][1:].copy()

    # Renombrar las columnas
    df_estacion.columns = ['DateTime', 'O3', 'CO', 'NO2', 'SO2', 'PM10', 'PM2.5']

    # Agregar columna con el nombre de la estación
    df_estacion['Estacion'] = estacion

    # Concatenar al DataFrame total
    df_total = pd.concat([df_total, df_estacion], ignore_index=True)

# Mostrar primeras filas del conjunto total
df_total
```

```python
# Guardar el DataFrame en un archivo CSV
df_total.to_csv('data/csv/2024.csv', index=False)
```
